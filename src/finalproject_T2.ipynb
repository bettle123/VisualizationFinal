```python
# import necessary packages used in this project
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a pandas DataFrame from the file
df_filename = pd.read_csv("stats/filenames.csv", header=None);
df_median = pd.read_csv("stats/tev/tev_median.csv");
df_mean = pd.read_csv("stats/tev/tev_mean.csv");
df_std = pd.read_csv("stats/tev/tev_std.csv");
df_size = pd.read_csv("stats/tev/tev_size.csv");
df_max = pd.read_csv("stats/tev/tev_max.csv");
df_min = pd.read_csv("stats/tev/tev_min.csv");
df_range = pd.read_csv("stats/tev/tev_range.csv");

#data
print(df_filename.head())
```

```python
# get the timestep from filename and remove the negative data because of numerical issues
timestep = [];
for i in range(0, df_filename.size):
    whole_str = df_filename.iloc[i,0];
    sub_str = whole_str[22:27];
    #print(sub_str);
    dtimestep = int(sub_str);
    timestep.append(dtimestep);
```

```python
# plot every timestep stats with 0
# median
plt.plot(timestep,df_median)
plt.xlabel('timestep')
plt.title("tev median")
plt.savefig("plots/series/tev/tev_median.png")
plt.show()

# mean
plt.plot(timestep,df_mean)
plt.xlabel('timestep')
plt.title("tev mean")
plt.savefig("plots/series/tev/tev_mean.png")
plt.show()


# std
plt.plot(timestep,df_std)
plt.xlabel('timestep')
plt.title("tev std")
plt.savefig("plots/series/tev/tev_std.png")
plt.show()

# size
plt.plot(timestep,df_size)
plt.xlabel('timestep')
plt.title("tev size")
plt.savefig("plots/series/tev/tev_size.png")
plt.show()

# max
plt.plot(timestep,df_max)
plt.xlabel('timestep')
plt.title("tev max")
plt.savefig("plots/series/tev/tev_max.png")
plt.show()

# min
plt.plot(timestep,df_min)
plt.xlabel('timestep')
plt.title("tev min")
plt.savefig("plots/series/tev/tev_min.png")
plt.show()

# range
plt.plot(timestep,df_range)
plt.xlabel('timestep')
plt.title("tev range")
plt.savefig("plots/series/tev/tev_range.png")
plt.show()
```
