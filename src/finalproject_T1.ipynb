```python
# import necessary packages used in this project
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a pandas DataFrame from the file
# data with 0
df_filename = pd.read_csv("stats/filenames.csv", header=None);
df_median = pd.read_csv("stats/v03_median.csv");
df_mean = pd.read_csv("stats/v03_mean.csv");
df_std = pd.read_csv("stats/v03_std.csv");
df_size = pd.read_csv("stats/v03_size.csv");
df_max = pd.read_csv("stats/v03_max.csv");
df_min = pd.read_csv("stats/v03_min.csv");
df_range = pd.read_csv("stats/v03_range.csv");

# data without 0
df_median_n0 = pd.read_csv("stats/v03_median_n0.csv");
df_mean_n0 = pd.read_csv("stats/v03_mean_n0.csv");
df_std_n0 = pd.read_csv("stats/v03_std_n0.csv");
df_size_n0 = pd.read_csv("stats/v03_size_n0.csv");
df_max_n0 = pd.read_csv("stats/v03_max_n0.csv");
df_min_n0 = pd.read_csv("stats/v03_min_n0.csv");
df_range_n0 = pd.read_csv("stats/v03_range_n0.csv");





#data
print(df_filename.head())
```

```python
# get the timestep from filename and remove the negative data because of numerical issues
timestep = [];
for i in range(0, df_filename.size):
    whole_str = df_filename.iloc[i,0];
    sub_str = whole_str[22:27];
    #print(sub_str);
    dtimestep = int(sub_str);
    timestep.append(dtimestep);
for i in range(0, df_min.size):
    if(df_min.iloc[i,0]<0.0):
        df_min.iloc[i,0] = 0.0;

for i in range(0, df_min_n0.size):
    if(df_min_n0.iloc[i,0]<0.0):
        df_min_n0.iloc[i,0] = 0.0;
#print(timestep);
```

```python
# plot every timestep stats with 0
# median
plt.plot(timestep,df_median)
plt.xlabel('timestep')
plt.title("v03 median")
plt.savefig("plots/series/w0/v03_median.png")
plt.show()

# mean
plt.plot(timestep,df_mean)
plt.xlabel('timestep')
plt.title("v03 mean")
plt.savefig("plots/series/w0/v03_mean.png")
plt.show()


# std
plt.plot(timestep,df_std)
plt.xlabel('timestep')
plt.title("v03 std")
plt.savefig("plots/series/w0/v03_std.png")
plt.show()

# size
plt.plot(timestep,df_size)
plt.xlabel('timestep')
plt.title("v03 size")
plt.savefig("plots/series/w0/v03_size.png")
plt.show()

# max
plt.plot(timestep,df_max)
plt.xlabel('timestep')
plt.title("v03 max")
plt.savefig("plots/series/w0/v03_max.png")
plt.show()

# min
plt.plot(timestep,df_min)
plt.xlabel('timestep')
plt.title("v03 min")
plt.savefig("plots/series/w0/v03_min.png")
plt.show()

# range
plt.plot(timestep,df_range)
plt.xlabel('timestep')
plt.title("v03 range")
plt.savefig("plots/series/w0/v03_range.png")
plt.show()


```

```python
# plot every timestep stats without 0
# median
plt.plot(timestep,df_median_n0)
plt.xlabel('timestep')
plt.title("v03 median n0")
plt.savefig("plots/series/n0/v03_median_n0.png")
plt.show()

# mean
plt.plot(timestep,df_mean_n0)
plt.xlabel('timestep')
plt.title("v03 mean n0")
plt.savefig("plots/series/n0/v03_mean_n0.png")
plt.show()


# std
plt.plot(timestep,df_std_n0)
plt.xlabel('timestep')
plt.title("v03 std n0")
plt.savefig("plots/series/n0/v03_std_n0.png")
plt.show()

# size
plt.plot(timestep,df_size_n0)
plt.xlabel('timestep')
plt.title("v03 size n0")
plt.savefig("plots/series/n0/v03_size_n0.png")
plt.show()

# max
plt.plot(timestep,df_max_n0)
plt.xlabel('timestep')
plt.title("v03 max n0")
plt.savefig("plots/series/n0/v03_max_n0.png")
plt.show()

# min
plt.plot(timestep,df_min_n0)
plt.xlabel('timestep')
plt.title("v03 min n0")
plt.savefig("plots/series/n0/v03_min_n0.png")
plt.show()

# range
plt.plot(timestep,df_range_n0)
plt.xlabel('timestep')
plt.title("v03 range n0")
plt.savefig("plots/series/n0/v03_range_n0.png")
plt.show()
```
