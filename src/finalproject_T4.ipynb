```python
# import necessary packages used in this project
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a pandas DataFrame from the file
df_filename = pd.read_csv("stats/yA31/filenames.csv", header=None);
df_median = pd.read_csv("stats/yA31/v02/v02_median_all.csv");
df_mean = pd.read_csv("stats/yA31/v02/v02_mean_all.csv");
df_std = pd.read_csv("stats/yA31/v02/v02_std_all.csv");
df_size = pd.read_csv("stats/yA31/v02/v02_size_all.csv");
df_max = pd.read_csv("stats/yA31/v02/v02_max_all.csv");
df_min = pd.read_csv("stats/yA31/v02/v02_min_all.csv");
df_range = pd.read_csv("stats/yA31/v02/v02_range_all.csv");

#data
print(df_filename.head())
print(df_median.head())
```

```python
# get the timestep from filename and remove the negative data because of numerical issues
timestep = [];
for i in range(0, df_filename.size):
    whole_str = df_filename.iloc[i,0];
    sub_str = whole_str[22:27];
    #print(sub_str);
    dtimestep = int(sub_str);
    timestep.append(dtimestep);
print(timestep)
print(df_median)
```

```python
# plot every timestep stats with 0
# median
plt.plot(timestep,df_median)
plt.xlabel('timestep')
plt.title("v02 median")
plt.savefig("plots/yA31/series/v02/v02_median.png")
plt.show()
# mean
plt.plot(timestep,df_mean)
plt.xlabel('timestep')
plt.title("v02 mean")
plt.savefig("plots/yA31/series/v02/_mean.png")
plt.show()


# std
plt.plot(timestep,df_std)
plt.xlabel('timestep')
plt.title("v02 std")
plt.savefig("plots/yA31/series/v02/v02_std.png")
plt.show()

# size
plt.plot(timestep,df_size)
plt.xlabel('timestep')
plt.title("v02 size")
plt.savefig("plots/yA31/series/v02/v02_size.png")
plt.show()

# max
plt.plot(timestep,df_max)
plt.xlabel('timestep')
plt.title("v02 max")
plt.savefig("plots/yA31/series/v02/v02_max.png")
plt.show()

# min
plt.plot(timestep,df_min)
plt.xlabel('timestep')
plt.title("v02 min")
plt.savefig("plots/yA31/series/v02/v02_min.png")
plt.show()

# range
plt.plot(timestep,df_range)
plt.xlabel('timestep')
plt.title("v02 range")
plt.savefig("plots/yA31/series/v02/v02_range.png")
plt.show()

```
