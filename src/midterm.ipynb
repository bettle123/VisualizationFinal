```python
# import necessary packages used in this project
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a pandas DataFrame from the file
df = pd.read_csv("data.csv");
#data
print(df.head())
```

```python
# summary table for all flights
summary = df.groupby(["ORIGIN", "DEST"]).size().reset_index(name='total_flights');
#print(summary);
#print(len(summary.index));
# There are 5842 kinds of flights in our data
DEP = summary.iloc[0,0];
ARR = summary.iloc[0,1];
total_flight = summary.iloc[0,2];
sub_data = df[(df['ORIGIN']==DEP)&(df['DEST']==ARR)];

AIR_TIME = sub_data['AIR_TIME'].sum()/total_flight;
DISTANCE = sub_data['DISTANCE'].sum()/total_flight;

DEP_delay_times = sub_data['DEP_DEL15'].sum();
DEP_Pr = DEP_delay_times/total_flight;
DEP_delay_time = sub_data['DEP_DELAY_NEW'].sum()/DEP_delay_times;

ARR_delay_times = sub_data['ARR_DEL15'].sum();
ARR_Pr = ARR_delay_times/total_flight;
ARR_delay_time = sub_data['ARR_DELAY_NEW'].sum()/ARR_delay_times;

print("DEP: ",DEP);
print("ARR: ",ARR);
print("total_flight: ", total_flight);
print("AIR_TIME: ",AIR_TIME);
print("DISTANCE: ",DISTANCE);
print("DEP_delay_times: ",DEP_delay_times);
print("DEP_Pr: ",DEP_Pr);
print("DEP_delay_time: ",DEP_delay_time);
print("ARR_delay_times: ",ARR_delay_times);
print("ARR_Pr: ",ARR_Pr);
print("ARR_delay_time: ",ARR_delay_time);

DEP = [];
ARR = [];
total_flight = [];
AIR_TIME = [];
DISTANCE = [];
DEP_delay_times = [];
DEP_Pr = [];
DEP_delay_time = [];
ARR_delay_times =[];
ARR_Pr = [];
ARR_delay_time = [];
#for i in range(0, 10):
for i in range(0, len(summary.index)):
    tmp_DEP = summary.iloc[i,0];
    tmp_ARR = summary.iloc[i,1];
    tmp_total_flight = summary.iloc[i,2];

    sub_data = df[(df['ORIGIN']==tmp_DEP)&(df['DEST']==tmp_ARR)];
    
    tmp_AIR_TIME = sub_data['AIR_TIME'].sum()/tmp_total_flight;
    tmp_DISTANCE = sub_data['DISTANCE'].sum()/tmp_total_flight;

    tmp_DEP_delay_time =0;
    tmp_ARR_delay_time =0;
    
    tmp_DEP_delay_times = sub_data['DEP_DEL15'].sum();
    tmp_DEP_Pr = tmp_DEP_delay_times/tmp_total_flight;
    if(tmp_DEP_delay_times>0):
        tmp_DEP_delay_time = sub_data['DEP_DELAY_NEW'].sum()/tmp_DEP_delay_times;
    if(tmp_DEP_delay_times==0):
        tmp_DEP_delay_time =0;
    tmp_ARR_delay_times = sub_data['ARR_DEL15'].sum();
    tmp_ARR_Pr = tmp_ARR_delay_times/tmp_total_flight;
    if(tmp_ARR_delay_times>0):
        tmp_ARR_delay_time = sub_data['ARR_DELAY_NEW'].sum()/tmp_ARR_delay_times;
    if(tmp_ARR_delay_times==0):
        tmp_ARR_delay_time =0;
    DEP.append(tmp_DEP);
    ARR.append(tmp_ARR);
    total_flight.append(tmp_total_flight);
    AIR_TIME.append(tmp_AIR_TIME);
    DISTANCE.append(tmp_DISTANCE);
    DEP_delay_times.append(tmp_DEP_delay_times);
    DEP_Pr.append(tmp_DEP_Pr);
    DEP_delay_time.append(tmp_DEP_delay_time);
    ARR_delay_times.append(tmp_ARR_delay_times);
    ARR_Pr.append(tmp_ARR_Pr);
    ARR_delay_time.append(tmp_ARR_delay_time);


dict = {
    "DEP":DEP,
    "ARR":ARR,
    "total_flight":total_flight,
    "AIR_TIME":AIR_TIME,
    "DISTANCE":DISTANCE,
    "DEP_delay_times":DEP_delay_times,
    "DEP_Pr":DEP_Pr,
    "DEP_delay_time":DEP_delay_time,
    "ARR_delay_times":ARR_delay_times,
    "ARR_PR":ARR_Pr,
    "ARR_delay_time":ARR_delay_time
}
final_df = pd.DataFrame(dict)    
print(final_df.head());
final_df.to_csv("summary_table.csv", index=False)
print(len(summary.index));
print(df.groupby(["ORIGIN"]).size());
```
